```python
df = df.convert_dtypes()
```

```python
#Change the data type
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.dtypes
```

```python
df.columns
```

```python
# Strip the string data
text_cols = df.select_dtypes(include= "string").columns
# text_cols
df[text_cols] = df[text_cols].apply(lambda x: x.str.strip())
```

```python
# Detect invalid values (e.g. age > 100)
df['customer_age'].describe()
```

**There is no invalid age values.**

```python
#Checking unique values 
cols = ['category', 'payment_method', 'region', 'returned', 'customer_gender']
for col in cols:
    print(f"\n{col}:")
    print(sorted(df[col].unique()))
```

```python
# Building profit column ( profit margin = total profit/ total revenue )
df['profit'] = df['total_amount'] * df['profit_margin']
```

```python
#Export Csv Files 
df.to_csv("ecommerce_sales_transactions_cleaned_dataset.csv",index=False)    
```
